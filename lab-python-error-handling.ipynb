````python
# Exercise 1:
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
````

````python
#Exercise 2:
def calculate_total_price(products):
    total_price = 0
    for product in products:
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price for {product}: "))
                if price < 0:
                    raise ValueError("Invalid price! Please enter a non-negative value.")
                valid_price = True
                total_price += price
            except ValueError as error:
                print(f"Error: {error}")
    return total_price
````

````python
#Exercise 3:

def get_customer_orders(inventory):
    orders = {}
    valid_order_number = False
    while not valid_order_number:
        try:
            order_count = int(input("How many products would you like to order? "))
            if order_count < 0:
                raise ValueError("Number of orders cannot be negative. Please enter a valid number.")
            valid_order_number = True
        except ValueError as error:
            print(f"Error: {error}")

    for _ in range(order_count):
        valid_product = False
        while not valid_product:
            try:
                product = input("Enter the product name: ")
                if product not in inventory:
                    raise ValueError(f"Invalid product name! {product} is not available in the inventory.")
                if inventory[product] <= 0:
                    raise ValueError(f"Sorry, {product} is out of stock.")
                valid_product = True

                # Now check for quantity
                valid_quantity = False
                while not valid_quantity:
                    try:
                        quantity = int(input(f"How many {product}s would you like to order? "))
                        if quantity < 0 or quantity > inventory[product]:
                            raise ValueError(f"Invalid quantity! Available stock for {product}: {inventory[product]}")
                        valid_quantity = True
                        orders[product] = quantity
                        inventory[product] -= quantity  # Update inventory
                    except ValueError as error:
                        print(f"Error: {error}")

            except ValueError as error:
                print(f"Error: {error}")

    return orders
````

````python
#Exercise 4:

def main():
    products = ["apple", "banana", "orange"]
    inventory = initialize_inventory(products)
    total_price = calculate_total_price(products)
    print(f"Total price of all products: ${total_price:.2f}")
    customer_orders = get_customer_orders(inventory)
    print(f"Customer orders: {customer_orders}")
    print(f"Updated inventory: {inventory}")

# Run the main program
main()
````
